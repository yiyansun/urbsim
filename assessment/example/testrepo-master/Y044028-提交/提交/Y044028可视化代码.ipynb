# 模型3

```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
initial_population = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
mean_sugar = [54.56524752, 47.58133663, 40.05409241, 38.90066832, 36.60285149, 33.21133663, 31.1419802, 28.43178218, 25.78448845, 23.94421782]
count_turtles = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

# Create main figure and axis
fig, ax1 = plt.subplots(figsize=(12, 7))

# Twin the axis for two different y-axes
ax2 = ax1.twinx()

# Plot data
ax1.plot(initial_population, mean_sugar, 'o-', color='blue', markerfacecolor='red', markersize=8, label='Mean Sugar Amount')
ax2.plot(initial_population, count_turtles, 's-', color='green', markerfacecolor='yellow', markersize=8, label='Count of Turtles')

# Set the y axis labels
ax1.set_ylabel('Mean Sugar Amount per Agent', color='blue', fontsize=14)
ax2.set_ylabel('Count of Turtles', color='green', fontsize=14)

# Enhancing the plot
ax1.set_xlabel('Initial Population', fontsize=14)
# ax1.set_title('Mean Sugar Amount and Turtle Count vs. Initial Population in Sugarscape 3 Wealth Distribution', fontsize=16)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
ax1.set_xticks(np.arange(min(initial_population), max(initial_population)+100, 100))
ax1.tick_params(axis='y', colors='blue', labelsize=14)
ax1.tick_params(axis='x', labelsize=14)
ax2.tick_params(axis='y', colors='green', labelsize=14)
ax1.legend(loc='upper left', fontsize=14)
ax2.legend(loc='upper right', fontsize=14)

# Show plot
plt.tight_layout()
plt.show()
```
